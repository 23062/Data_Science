# load Dataset's

### 1. wine dataset

```python
from sklearn import datasets
import pandas as pd

wine = datasets.load_wine()

wine_df = pd.DataFrame(wine.data, columns = wine.feature_names)
wine_df["WineType"] = [wine.target_names[wine_typ] for wine_typ in wine.target]
wine_df["WineTypeInt"] = wine.target

wine_df.head()
```

### 2. Nifty50 dataset

```python
import pandas as pd

nifty50_df = pd.read_excel("H:\\Analyst projects\\excel-data\\nifty50.xlsx",)

nifty50_df = nifty50_df.reset_index()
nifty50_df.head()
```

#### scatter chart

```python
wine_df.plot(x='alcohol', y='malic_acid',kind='scatter')
```

```python
wine_df.plot.scatter(x='alcohol', y='malic_acid', 
                     xlabel = 'Alcohol', ylabel = 'Malic Acid', title = 'Alcohol VS Malic Acid',
                    c= 'red', figsize=(10,7))
```

```python
wine_df.plot.scatter(x='alcohol', y='malic_acid', 
                     xlabel = 'Alcohol', ylabel = 'Malic Acid', title = 'Alcohol VS Malic Acid',
                    c= 'WineTypeInt', cmap='viridis', marker ='s', figsize=(10,7))
```

#### bar graph


```python
avg_wine_df = wine_df.groupby('WineType').mean().reset_index()
avg_wine_df
```

```python
avg_wine_df.plot.bar(x='WineType',y='flavanoids' , 
                     xlabel = 'WneType', ylabel= 'flavanoids', title='WineType VS Flavanoids',
                    color = 'green', figsize=(5,5))
```

#### Horizonal bar graph

```python
avg_wine_df.plot.barh(x='WineType',y='flavanoids' , 
                     xlabel = 'WneType', ylabel= 'flavanoids', title='WineType VS Flavanoids',
                    color = 'yellow', figsize=(5,5))
```

#### Grouped Bar Chart

```python
ingredients = ['malic_acid','ash','total_phenols','flavanoids','proanthocyanins','hue']

avg_wine_df.plot.bar(x = 'WineType', y =ingredients, figsize=(10,7),
                    xlabel='WineType', ylabel='Ingredients', title="Average of WineType VS Ingredients")
```

#### stacked Bar chart

```python
ingredients = ['malic_acid','ash','total_phenols','flavanoids','proanthocyanins','hue']

avg_wine_df.plot.bar(x = 'WineType', y =ingredients, figsize=(10,7), 
                     stacked = True,
                    xlabel='WineType', ylabel='Ingredients', title="Average of WineType VS Ingredients")
```

### 5. Pie chart

```python
wine_cnt = wine_df.groupby('WineType').count()[['alcohol']].rename(columns={"alcohol":"count"})
wine_cnt
```

```python
wine_cnt.plot.pie(y='count' ,  figsize=(8,8),
                  autopct='%0.2f%%',
                  title='wine type  distribution')
```

```python
wine_cnt.plot.pie(y='count' ,  figsize=(5,5),
                  autopct='%0.2f%%', 
                  # to change the color and font size of the chart
                  textprops=dict(fontsize=15, fontweight='bold', color='white'),
                  title='wine type  distribution')
```

#### donut chart

```python
wine_cnt.plot.pie(y='count' ,  figsize=(5,5),
                  autopct='%0.2f%%', textprops=dict(fontsize=15, fontweight='bold', color='white'),
                   title='wine type  distribution',
                  # this code for donut chart and range between 0 to 1
                  wedgeprops=dict(width=0.7)
                 )
```
